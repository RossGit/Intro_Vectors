```python
#1
import numpy as np

u = np.array([1,-2,4])
v = np.array([3,5,1])
w = np.array([2,1,-3])

print("Given the following vectors:")
print("u : ",u)
print("v : ",v)
print("w : ",w)
print("Find: ")
print("a.) 3ùêÆ-2ùêØ: ", (3*u) - (2*v))
print("b.) 5ùêÆ+3ùêØ-4ùê∞: ", (5*u) + (3*v) - (4*w))
print('c.) u.v: ', u.dot(v))
print('c.) v.w: ', v.dot(w))
print('c.) u.w: ', u.dot(w))
print('d.) ||u||: ', np.sqrt(u.dot(u)))
print('d.) ||v||: ', np.sqrt(v.dot(v)))
print('d.) ||w||: ', np.sqrt(w.dot(w)))
print("e.) angle theta between u and v:",u.dot(v)/(np.linalg.norm(u)*np.linalg.norm(v)))
print("f.) angle alpha between v and w:",v.dot(w)/(np.linalg.norm(v)*np.linalg.norm(w)))
print("g.) angle theta between u and w:",u.dot(w)/(np.linalg.norm(u)*np.linalg.norm(w)))
print("h.) d(u,v): ", np.linalg.norm(u-v))
print("i.) d(w,v): ", np.linalg.norm(w-v))
print("j.) d(u,w): ", np.linalg.norm(u-w))
print("k.) proj(u,v): ",u.dot(v)*v/(np.linalg.norm(v)**2) )
print("l.) proj(w,v): ",w.dot(v)*v/(np.linalg.norm(v)**2) )
print("k.) proj(u,w): ",u.dot(w)*w/(np.linalg.norm(w)**2) )

```

```python
#5
import numpy as np

u = np.array([1,2,-2])
v = np.array([3,-12,4])
k = -3

print("Given the following vectors and scalar: ")
print("u : ",u)
print("v : ",v)
print("k = ",k)
print("Find the following: ")
print("||u||, ||v||, ||u+v||, ||ku||")
print("||u|| : " , np.sqrt(u.dot(u)))
print("||v|| : " , np.sqrt(v.dot(v)))
print("||u+v|| : " , np.sqrt((u+v).dot(u+v)))
print("||ku|| : " , np.sqrt((u.dot(k)).dot(u.dot(k))))

```
